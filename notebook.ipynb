# Recuperation de secteur d'entreprise

```python


def getSector(URL_input = "https://www.goafricaonline.com/ci/annuaire"):

    print('Fetching from url =', URL_input)
    URLdata = requests.get(URL_input)
    soup = bs4.BeautifulSoup(URLdata.text, "html.parser")
    a_tags = soup.find_all('a', class_='stretched-link text-center')
     
    links = []
    sectors = []
    sector = None
    link = None
    print('Please wait..')
    print('ok  ',a_tags)
    for a_tag in a_tags:
        try:
            link = a_tag['href']
            sector = a_tag.text
            links.append(link)
            sectors.append(sector)
            # print(link)
            # print(sector)
      
        except:
            pass
    print('\t\t\t Downloaded Successfully..\t\t ')
    print("nombre de linsk est {} et le nombre de sector est {} ".format(len(links), len(sectors)))
    return links, sectors
```

# Recuperation de sous seteurs

```python
def getSubSector(Sectors_df): 

        i=0
        for sector, link in zip(Sectors_df.secteur, Sectors_df.lien):
                sublink,subsect =  getSector(link)
                sectors = [ sector for i in subsect ]
                Sectors_df = None
                i += 1
                if i ==1:
                        Sectors_df = pd.DataFrame({'secteur' : sectors, 'sous_secteur':subsect, 'sous_lien': sublink})
                        Sectors_df.to_csv("data/Sous_Secteur.csv", index=False,header=True)
                else:
                        Sectors_df = pd.DataFrame({'secteur' : sectors, 'sous_secteur':subsect, 'sous_lien': sublink})
                        Sectors_df.to_csv("data/Sous_Secteur.csv", index=False,mode = 'a', header=False)
                



        print(Sectors_df)
```
